```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load CSV with BOM fix
df = pd.read_csv("worldpopulationdata.csv", encoding='utf-8-sig')

# Clean column names
df.columns = df.columns.str.strip()

# Quick peek at columns
print(df.columns.tolist())

# Select relevant columns: country and 2022 population
df_2022 = df[['Country Name', '2022']].dropna()

# Sort top 10 by population
top_10 = df_2022.sort_values(by='2022', ascending=False).head(10)

# Plot bar chart - Top 10 most populated countries in 2022
plt.figure(figsize=(12,6))
sns.barplot(data=top_10, x='Country Name', y='2022', palette='magma')
plt.title("Top 10 Most Populated Countries in 2022")
plt.xticks(rotation=45)
plt.ylabel("Population")
plt.xlabel("Country")
plt.tight_layout()
plt.show()

# Plot histogram of population distribution in 2022
plt.figure(figsize=(10,6))
sns.histplot(data=df_2022, x='2022', bins=30, kde=True, color='teal')
plt.title("Distribution of Country Populations in 2022")
plt.xlabel("Population")
plt.ylabel("Number of Countries")
plt.tight_layout()
plt.show()
```
