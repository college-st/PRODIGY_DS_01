```python
import pandas as pd

url = "https://raw.githubusercontent.com/ved-stack/PRODIGY_DS_03/main/bank-additional-full.csv"
df = pd.read_csv(url, sep=';')

print(df.shape)
df.head()
```

```python
print(df.shape)
df.info()
df['y'].value_counts()
df.describe(include='all')
```

```python
df.drop('duration', axis=1, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

df_enc = df.copy()
le = LabelEncoder()
for col in df_enc.select_dtypes('object').columns:
    df_enc[col] = le.fit_transform(df_enc[col])
```

```python
from sklearn.model_selection import train_test_split

X = df_enc.drop('y', axis=1)
y = df_enc['y']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_train_bal, y_train_bal = smote.fit_resample(X_train, y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier(
    max_depth=5,
    min_samples_split=20,
    class_weight='balanced',
    random_state=42
)
clf.fit(X_train_bal, y_train_bal)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
print("AUC:", roc_auc_score(y_test, clf.predict_proba(X_test)[:,1]))
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 10))
plot_tree(clf, feature_names=X.columns, class_names=['no', 'yes'], filled=True)
plt.show()
```

```python
feat_imp = pd.Series(clf.feature_importances_, index=X.columns).sort_values(ascending=False)
affordance = feat_imp.head(10)
print(affordance)
```
