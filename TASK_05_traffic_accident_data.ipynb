```python
import pandas as pd
df = pd.read_csv('US_Accidents_Sample.csv')
df.head()
```

```python
# Summary of dataframe info (columns, non-null counts, datatypes)
df.info()

# Summary statistics for numeric columns
df.describe()
```

```python
df['Severity'].value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot distribution of accident severity
sns.countplot(x='Severity', data=df)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))
sns.countplot(y='Weather_Condition', data=df, order=df['Weather_Condition'].value_counts().index)
plt.title('Accidents by Weather Condition')
plt.show()
```

```python
road_features = ['Amenity', 'Bump', 'Crossing', 'Give_Way', 'Junction', 'No_Exit',
                 'Railway', 'Roundabout', 'Station', 'Stop', 'Traffic_Calming',
                 'Traffic_Signal', 'Turning_Loop']

road_feature_counts = df[road_features].sum().sort_values(ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(x=road_feature_counts.values, y=road_feature_counts.index)
plt.title('Accidents by Road Features')
plt.xlabel('Number of Accidents')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Sunrise_Sunset', data=df)
plt.title('Accidents by Day/Night')
plt.show()
```
